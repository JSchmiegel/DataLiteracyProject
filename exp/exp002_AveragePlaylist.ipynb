```python
dataset = get_data_tracks(pre=True)
```

```python
print(dataset['track_uri'].nunique())
```

```python
count_unique_apperance_tracks = dataset['track_uri'].value_counts().to_numpy()
#print(count_unique_apperance_tracks.head(100))
for entry in count_unique_apperance_tracks:
    print(entry)
    #if count_unique_apperance_tracks > 1:
    #    print(count_unique_apperance_tracks)
    
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(count_unique_apperance_tracks)
axs[1].plot(count_unique_apperance_tracks)
axs[1].set_xscale("log")

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].hist(count_unique_apperance_tracks)
axs[1].hist(count_unique_apperance_tracks, bins=np.logspace(np.log10(1),np.log10(10)))
axs[1].set_xscale("log")

plt.show()
```
