```python
import numpy as np
import pandas as pd

from wordcloud import WordCloud

from matplotlib import pyplot as plt
```

```python
data_path = 'csv_data/'
THREE_K = '_3k'
TWENTY_K = '_20k'
HUNDRED_K = '_100k'
```

```python
size = TWENTY_K
```

```python
df = pd.read_csv(data_path + 'playlist_df' + size + '.csv', index_col=0)
```

```python
name_list = ' '.join(df['name'].tolist()).split()
name_list = [x.lower() for x in name_list]
```

```python
unique, counts = np.unique(np.array(name_list), return_counts=True) 
df_name_counts = pd.DataFrame([unique, counts]).T
df_sorted = df_name_counts.sort_values(by=1)[::-1][:70]
```

```python
words_dict = dict(zip(df_sorted[0], df_sorted[1]))
```

```python
wordcloud = WordCloud(width = 1000, height = 500,
                      colormap='copper', mode = 'RGBA', background_color=None,
                      relative_scaling=0.3, prefer_horizontal=0.95).generate_from_frequencies(words_dict)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
